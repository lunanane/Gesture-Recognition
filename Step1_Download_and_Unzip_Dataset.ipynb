**Check for data integrity for both training and validation files**

If it extracted all correctly - e.g. now folders missing in the integrity check - then lets move on to the actual training[Linktext](https://)

```python
path_toCSV = '/content/drive/MyDrive/gesture_recognition/3D-CNN-Gesture-recognition/data_csv/20bn-jasper-subset-training-swipe.csv' #@param{type: 'string'}

path_dataset_train = '/content/drive/MyDrive/gesture_recognition/data-set-full/extracted/Train' #@param{type: 'string'}

label = pd.read_csv(path_toCSV,header=None,sep = ";")
label.head()
print(label)
targets_name = label[0].tolist()
targets_classification = label[1].tolist()
print(targets_name)
print(targets_classification)

counter_exist = 0
counter_missing = 0
max_inter = len(targets_name)
for i in range(0, max_inter, 1):
  check_folder = path_dataset_train + "/" + str(targets_name[i])
  if os.path.isdir(check_folder):
    counter_exist += 1
  else: 
    counter_missing += 1

  print("existing folder:" + str(counter_exist) + " / missing folder: " + str(counter_missing) + "/ total folders: " + str(len(targets_name)))

```
